### Importing the necessary libraries and exploring the dataset

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data = pd.read_csv("SampleSuperstore.csv")
data.head(5)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates()
```

```python
data.shape
```

```python
data1 = data.drop(['Postal Code' ] , axis =1)
data1= data.drop(['Country'] , axis =1)
data1
```

```python
data1.describe()
```

```python
corr1 = data1.corr()
corr1
```

```python
plt.figure(figsize = (12,6))
sns.heatmap(corr1 , annot = True)
plt.show()
```

### Data Visualization

```python
data1['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category', 'Category', data = data1)
plt.show()
```

```python
print(data1 ['Ship Mode'].value_counts())
plt.figure(figsize = (10,6))
sns.countplot(x= data1['Ship Mode'])
plt.show()
```

```python
print(data1['Region'].value_counts())
plt.figure(figsize= (15,6))
sns.countplot(x= data1['Region'])
plt.show()
```

```python
print(data1['Category'].value_counts())
plt.figure(figsize= (15,6))
sns.countplot(x= data1['Category'])
plt.show()
```

```python
plt.figure(figsize = (12,6))
plt.pie(data1['Category'].value_counts(), labels = data1['Category'].value_counts().index , startangle = 180, radius = 1)
plt.title('Main category of operation')
plt.show()
print(data1['Category'].value_counts())
```

```python
plt.figure(figsize = (12,6))
plt.pie(data1['Sub-Category'].value_counts(), labels = data1['Sub-Category'].value_counts().index , startangle = 180, radius = 1)
plt.show()
print(data1['Sub-Category'].value_counts())
```

## Analysis of Subcategory wrt Profit

```python
plt.figure(figsize = (12,6))
sns.barplot(x=data1['Sub-Category'], y = data1['Profit'])
plt.xticks(rotation = 90)
plt.show()
```
